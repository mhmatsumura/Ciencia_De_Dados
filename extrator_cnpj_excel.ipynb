#O que este código faz (Algoritmo):

1. percorre a coluna A

2. Caso comece com "INSC.EST.:" atualiza o valor de "cnpj"

3. Caso seja um número adiciona o valor de cnpj na coluna H da respectiva linha.

```python
from google.colab import files
import pandas as pd
import io

arquivo_baixado = files.upload() # tela escolher arquivo
nome_arquivo = next(iter(arquivo_baixado)) #captura o nome do arquivo escolhido pelo usuario

df_fonte = pd.read_excel(io.BytesIO(arquivo_baixado[nome_arquivo])) # transfere os dados do arquivo do excel para o dataframe "df_fonte"

print(df_fonte.columns) #imprimindo o nome das colunas geradas
```

```python
# renomeando as colunas e imprimindo
df_fonte.rename(columns={'Unnamed: 0':'CLASS_FISCAL','Unnamed: 1':'DISCRIMINAÇÃO','Unnamed: 2':'UNID','Unnamed: 3':'QUANTIDADE','Unnamed: 4':'UNITÁRIO','Unnamed: 5':'PARCIAL','Unnamed: 6':'TOTAL','Unnamed: 7':'CNPJ'}, inplace=True)
print(df_fonte.columns)
```

```python
cnpj = "" #iniciando a variável cnpj com vazio
df_resultado = pd.DataFrame() #criando o dataframe df_resultado

for index, row in df_fonte.iterrows(): #percorrendo toda a planilha linha por linha
  colunaA = str(row["CLASS_FISCAL"]) # captura o valor contido na linha atual na coluna "CLASS_FISCAL"
  if colunaA.startswith('INSC.EST.:'): # se o valor iniciar com 'INSC.EST.:'
    cnpj = colunaA[colunaA.index("C.N.P.J.  :")+11:].strip() # segmenta a string eliminado até o ":"" do "C.N.P.J.  :",elimina espaços em branco
  elif colunaA.isnumeric(): #senão, se o valor for um número
    df_fonte.loc[index, "CNPJ"] = cnpj #seta o valor cnpj na coluna CNPJ na linha atual
    df_resultado = df_resultado.append(df_fonte.loc[index], ignore_index = True) #adiciona a linha alterada do df_fonte no df_resultado
   

df_resultado = df_resultado.astype({"CLASS_FISCAL": int, "QUANTIDADE": int}) # seta o tipo de dado para a coluna "CLASS_FISCAL" e "QUANTIDADE" como inteiro
df_resultado.head(10) #imprime as 10 primeiras ocorrências do dataframe df_resultado
```

```python
# Cria o arquivo do excel a partir do dataframe df_resultado 
df_resultado.to_excel('extrator_cnpj_resultado.xlsx')
# Faz o download do arquivo gerado
files.download('extrator_cnpj_resultado.xlsx')
```
