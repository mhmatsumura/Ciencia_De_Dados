```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

X_train = (df_total_treino.iloc[:,1:].values).astype('float32') # all pixel values
y_train = df_total_treino.iloc[:,0].values.astype('int32') # only labels i.e targets digits
X_test = df_total_teste.values.astype('float32')
X_train
```

```python
from sklearn.model_selection import train_test_split

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
# batches = gen.flow(X_train, y_train, batch_size=64)
# val_batches=gen.flow(X_val, y_val, batch_size=64)
```

```python
import numpy as np
mean_px = X_train.mean().astype(np.float32)
std_px = X_train.std().astype(np.float32)

def normaliza(x): 
    return (x-mean_px)/std_px
```

```python
X_train, X_test  = normaliza(X_train),  normaliza(X_test)

X_train_split, X_val_split  = normaliza(X_train_split),  normaliza(X_val_split)
```

```python
# Data Augmentation

from tensorflow.keras.preprocessing.image import ImageDataGenerator

gen =ImageDataGenerator(rotation_range=8, width_shift_range=0.08, shear_range=0.3,
                               height_shift_range=0.08, zoom_range=0.08)

# Define data augmentation parameters
gen2 = ImageDataGenerator(
    rotation_range=20,      # Random rotation between -20 and 20 degrees
    width_shift_range=0.2,  # Random horizontal shift by up to 20% of the image width
    height_shift_range=0.2, # Random vertical shift by up to 20% of the image height
    shear_range=0.2,        # Random shearing transformation
    zoom_range=0.2,         # Random zooming in/out by up to 20%
    # horizontal_flip=True,   # Randomly flip images horizontally
    # fill_mode='nearest'     # Fill points outside the boundaries with the nearest available values
)
```

```python
from tensorflow.keras.layers import Convolution2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dropout


# Definindo a arquitetura da rede neural
model = Sequential([

    # Convolution2D(16,(3,3), activation='relu'),
    # BatchNormalization(axis=1),
    # Convolution2D(16,(3,3), activation='relu'),
    # MaxPooling2D(),
    Convolution2D(32,(3,3), activation='relu'),
    BatchNormalization(axis=1),
    Convolution2D(32,(3,3), activation='relu'),
    MaxPooling2D(),
    Convolution2D(64,(3,3), activation='relu'),
    BatchNormalization(axis=1),
    Convolution2D(64,(3,3), activation='relu'),
    MaxPooling2D(),
    # Convolution2D(128,(3,3), activation='relu'),
    # BatchNormalization(axis=1),
    # Convolution2D(128,(3,3), activation='relu'),
    # MaxPooling2D(),

   
   
    Flatten(input_shape=(28, 28)),  # Camada para transformar a imagem em vetor
    # Dropout(0.1),
    # Dense(512, activation='relu'),   # Camada totalmente conectada com ativação ReLU
    # Dense(512, activation='relu'),   # Camada totalmente conectada com ativação ReLU
    # Dense(256, activation='relu'),   # Camada totalmente conectada com ativação ReLU
    Dense(128, activation='relu'),   # Camada totalmente conectada com ativação ReLU
    Dense(10, activation='softmax')  # Camada de saída com ativação Softmax para classificação
])

 # model = Sequential([
 #        Lambda(standardize, input_shape=(28,28,1)),
 #        Convolution2D(32,(3,3), activation='relu'),
 #        Convolution2D(32,(3,3), activation='relu'),
 #        MaxPooling2D(),
 #        Convolution2D(64,(3,3), activation='relu'),
 #        Convolution2D(64,(3,3), activation='relu'),
 #        MaxPooling2D(),
 #        Flatten(),
 #        Dense(512, activation='relu'),
 #        Dense(10, activation='softmax')
 #        ])



# Compilando o modelo

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```
